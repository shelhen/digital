```python
import pandas as pd
import warnings
from utils import drop_st,check_na, fillna__
warnings.filterwarnings('ignore')
pd.options.io.excel.xlsx.writer = 'xlsxwriter'

# 财务绩效处理
fin = r".\datas\全国上市企业财务绩效汇总表.csv"
dataset_fin = pd.read_csv(fin, index_col=None, header=0, dtype={'股票代码':'object'})  # , parse_dates=['上市日期']

# 删除st企业
dataset_fin = drop_st(dataset_fin, '股票简称')
losts = check_na(dataset_fin, rate=0.25)  # 删除缺失值太多的指标
dataset_fin.drop(losts, axis=1, inplace=True)
dataset_fin = fillna__(dataset_fin, '股票代码', rate=0.5)
drops = set(dataset_fin.loc[dataset_fin.isnull().values]['股票代码'].tolist())
dataset_fin = dataset_fin[~dataset_fin['股票代码'].isin(drops)]
dataset_fin.sort_values(['股票代码','截止年份'])
```

```python
# 读取非财务绩效数据
unfin = r".\datas\全国上市企业非财务绩效汇总表.csv"
# parse_dates=['上市日期']
dataset_unfin = pd.read_csv(unfin, index_col=None, header=0, dtype={'股票代码':'object'})
dataset_unfin = dataset_unfin[dataset_unfin['股票代码'].isin(dataset_fin['股票代码'].tolist())]
losts = check_na(dataset_unfin, rate=0.25)  # 删除缺失值太多的指标
dataset_unfin.drop(losts, axis=1, inplace=True)
dataset_unfin = fillna__(dataset_unfin, '股票代码', rate=0.5)
dataset_unfin.sort_values(['股票代码','截止年份'])
```

```python
# 缩尾处理
import numpy as np
from scipy.stats.mstats import winsorize

dataset = pd.merge(dataset_fin,dataset_unfin, on=['股票代码','截止年份'], how='outer')
# 清两职位合一和股权性质为0-1变量
dataset['两职合一'].fillna(0, inplace=True)
dataset['股权性质'].fillna(0, inplace=True)
dataset['两职合一'] = dataset['两职合一'].astype('int32')
dataset['股权性质'] = dataset['股权性质'].astype('int32')
# dataset.dropna(subset=['上市省份'],inplace=True)
dataset = fillna__(dataset, '股票代码', rate=0.3)
dataset.dropna(inplace=True)
dataset.sort_values(['股票代码','截止年份'],inplace=True)
dataset.rename(columns={'截止年份':'截止日期'}, inplace=True)
# dataset['上市日期'] = pd.to_datetime(dataset['上市日期'])
# dataset['上市日期'] = dataset['上市日期'].dt.year
# dataset['企业年限'] = dataset['截止日期'] - dataset['上市日期']
# dataset.drop(['上市日期','股票简称'], axis=1, inplace=True)
dataset.drop(['股票简称'], axis=1, inplace=True)
print(dataset.columns.to_list())


keywords = dataset.columns.drop(['股票代码', '截止日期', '行业代码','行业名称'])  # '上市省份',
for keyword in keywords:
    # 此处判断是否空值，是的话为空，否的话进行屏蔽空值和无效值的1%和99%缩尾处理
    dataset[keyword] =np.where(dataset[keyword].isnull(), np.nan, winsorize(np.ma.masked_invalid(dataset[keyword]),limits=(0.01,0.01)))

work_by = dataset.groupby('行业代码').count()
work_type = work_by.index.tolist()
work_code_dic = {work_type[i]:(i+1) for i in range(len(work_type))}
dataset['行业类别'] = dataset['行业代码'].apply(lambda x:work_code_dic[x])
# dataset_.drop('行业代码',axis=1, inplace=True)
dataset.sort_values(['股票代码','截止日期'],inplace=True)
dataset.reset_index(inplace=True, drop=True)
# 中国上市制造企业汇总表.txt
dataset.to_csv(r'.\datas\上市企业绩效汇总表.csv',encoding='utf8',index=None)
dataset
```

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
supply = pd.read_excel(r'C:\datas\企业信息数据\上市公司数据.xlsx',dtype={'id_str':'object'})
# dataset.head(10)
supply = supply[['id_str','year','LISTINGSTATE_11','firm_age','Sicda_str','EquityNature_p','prov_off','city_off','B001100000','lnSale','size','roa','roe']]
supply.dropna(subset=['id_str','Sicda_str'],inplace=True)

supply['year'] = supply['year'].apply(lambda x:int(x))
supply['Sicda_str'] = supply['Sicda_str'].apply(lambda x: int(x[1:]))
supply['EquityNature_p'] = supply['EquityNature_p'].apply(lambda x:1 if x=='国企' else 0)
supply=supply[(supply['year']>2006)&(supply['firm_age']>=0)&(supply['Sicda_str']>12)&(supply['Sicda_str']<44)]
supply['LISTINGSTATE_11'] =supply['LISTINGSTATE_11'].fillna('*ST')
supply['firm_age'] = supply['firm_age'].apply(lambda x:int(x))
supply.rename(columns={'id_str':'股票代码','year':'截止日期','LISTINGSTATE_11':'上市状态','firm_age':'企业年限','Sicda_str':'行业代码','EquityNature_p':'股权性质','prov_off':'上市省份','city_off':'上市城市','B001100000':'营业总收入','lnSale':"ln(营业收入)",'size':'ln(总资产)','roa':'总资产净利润率','roe':'净资产收益率'},inplace=True)
supply.loc[supply['股票代码']=='002532','上市省份']='浙江省'
supply.loc[supply['股票代码']=='002532','上市城市']='台州市'
supply.loc[supply['股票代码']=='600197','上市城市']='伊犁哈萨克自治州'
supply.loc[supply['股票代码']=='600075','上市城市']='伊犁哈萨克自治州'
supply.loc[supply['股票代码']=='600531','上市城市']='焦作市'
supply.drop(['行业代码','总资产净利润率','净资产收益率'],axis=1,inplace=True)
supply
```

```python
dataset_ = pd.merge(dataset, supply, on=['股票代码','截止日期'], how='left')
dataset_['上市状态'] =dataset_['上市状态'].fillna('ST')
dataset_ = dataset_[~dataset_['上市状态'].str.contains('ST')]
dataset_.dropna(inplace=True)
dataset_.sort_values(by=['股票代码','截止日期'],inplace=True)
dataset_.reset_index(inplace=True, drop=True)
dataset_.to_csv('./datas/中国数字化转型与企业绩效数据汇总.csv',encoding='utf8',index=None)
dataset_
```
