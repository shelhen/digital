```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')


dataset = pd.read_csv('./data/dataset.csv',dtype={'股票代码':'object'})
with open('./data/stcodes.txt','r',encoding='utf8') as f:
    scids = [scid for scid in f.read().split('\n') if scid!='']
dataset.rename(columns={'股权集中度4(%)':"股权集中度",'上市期限':'企业年限'},inplace=True)
for ns in ['两职合一','股权性质','企业年限']:
    dataset[ns] = dataset[ns].astype(int)
print(dataset.columns.tolist())
# dataset = dataset[dataset['行业代码']!=39]
# data.sort_values(['行业代码','截止日期'],inplace=True)

dataset = dataset[(dataset['截止日期']>2013)].set_index(['行业代码','截止日期'])
data = dataset[~dataset['股票代码'].isin(scids)]
data
```

```python
from linearmodels.panel import PanelOLS
# import statsmodels.api as sm

# ['营运能力', '行业代码', 
# '董事会规模', '两职合一', '股权集中度', '独董比例', '资产负债率', '资本密集度', '股权性质', '企业年限', '企业规模', '上市省份'

# '成本费用利润率', '销售期间费用率', '当年独立申请的发明数量', '当年联合申请的发明数量', 

# '当年独立获得的发明数量', '当年联合获得的发明数量', '当年独立申请专利总和', '当年联合申请专利总和', '当年独立获得专利总和', '当年联合获得专利总和', '独立申请实用新型&外观设计', '联合申请实用新型&外观设计', '独立获得实用新型&外观设计', '联合获得实用新型&外观设计'
# 'wufei', 'zhaochenyv', 'yuanchun', '无形资产数字化', 'csmar', 'lishouxi']

# 是否删除全部scid企业？
print(f'样本量{data.shape[0]}')
digital='wufei'
grant='当年独立申请专利总和'

controls = '股权集中度 + 企业年限 + 资本密集度'
#  + 资产负债率
# + 股权性质
# + 资本密集度
# + 董事会规模
# + 独董比例
# + 两职合一
#  + 企业规模 
# result = pd.DataFrame(columns=['Parameter','Std. Err.','T-stat','P-value','Lower CI','Upper CI'])

print(f'中介效应：{digital}——>{"营运能力"}')
model = PanelOLS.from_formula(f"营运能力 ~ 1+{digital}+{controls}+EntityEffects+TimeEffects", data=data)
res = model.fit(cov_type="clustered", cluster_entity=True)
result = pd.DataFrame(columns=['Parameter','P-value'])
result['Parameter'] = res.params
result['P-value'] = res.pvalues.round(5)
print(result)

print(f'中介效应：{digital}——>{grant}')
model = PanelOLS.from_formula(f"{grant} ~ 1+{digital}+{controls}+EntityEffects+TimeEffects", data=data)
res = model.fit(cov_type="clustered", cluster_entity=True)
result = pd.DataFrame(columns=['Parameter','P-value'])
result['Parameter'] = res.params
result['P-value'] = res.pvalues.round(5)
print(result)

print(f'中介效应：{digital}+{grant}——>{"营运能力"}')
model = PanelOLS.from_formula(f"营运能力~1+{digital}+{grant}+{controls}+EntityEffects+TimeEffects", data=data)
res = model.fit(cov_type="clustered", cluster_entity=True)
result = pd.DataFrame(columns=['Parameter','P-value'])
result['Parameter'] = res.params
result['P-value'] = res.pvalues.round(5)
print(result)
```
