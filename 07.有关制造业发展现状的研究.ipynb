```python
import numpy as np
import pandas as pd



# 绘制我国高端制造业上市公司总市值、收入和净利润变化趋势
assets = [0, 0, 0, 0, 0, 135555, 161873, 192927, 229342, 256929]
profit = [0, 0, 0, 0, 0, 3796, 5805, 8012, 8816, 7983]
revenue = [0, 0, 0, 0, 0, 82377, 93407, 116272, 134678, 146606]
market_value = [0, 0, 0, 0, 0, 160280, 273078, 353341, 288463, 291051]
year = [2014+i for i in range(10)]

dataset = pd.DataFrame([assets,profit,revenue,market_value], columns=year, index=['总资产(亿元)','净利润(亿元)','总收入(亿元)','总市值(亿元)']).T
dataset 
```

```python
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号


fig = plt.figure(figsize=(12, 6), dpi=300)
ax = fig.add_axes([0, 0, 1, 1])

colors=['#802827','#BB9A71']
width = 0.35
multiplier = 0
ax.set_title('1994-2023中国旅游市场规模及人均花费变化情况', fontsize=22)
ax.set_ylim([0, 70])
data = pd.DataFrame()
data['市场规模(亿元)'] = dataset['市场规模(亿元)'].apply(lambda x :round(x/1000,2))
data['国内旅游人次(万人)'] = dataset['国内旅游人次(万人)'].apply(lambda x:round(x/10000,2))
data.columns=['市场规模(千亿元)','国内旅游人次(亿人)']
i=0
for attribute, measurement in data.items():
    offset = width * multiplier
    rects = ax.bar(data.index+offset, measurement, width, label=attribute,alpha=0.9,color=colors[i])
    ax.bar_label(rects, padding=3)
    multiplier += 1
    i+=1
    
ax.set_xticks(dataset.index)
ax.set_xticklabels(dataset.index, rotation=40, fontsize=12)
ax2 = ax.twinx()
ax2.set_ylim([0, 1100])
ax2.plot(dataset['人均花费(元)'], color='#16171B',marker='o', linewidth=1.2, markersize=6, label='人均花费(元)')
ax.legend(loc=(0.30,0.9),ncols=2, frameon=False)
ax2.legend(loc=(0.60,0.9), frameon=False)

ax.spines["top"].set_alpha(0)
ax2.spines["top"].set_alpha(0)
# plt.savefig(os.path.join(picture_path, '中国旅游市场发展现状图.png'), dpi=300,bbox_inches='tight', pad_inches=0.0,transparent=True)
plt.show()
```
